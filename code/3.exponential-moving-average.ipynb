```python
import pandas as pd
import numpy as np
import mplfinance as mpf

df = pd.read_csv("./BTC-USD.csv", index_col='Date', parse_dates=True)
df
```

```python
# manual implementration
def calc_ema(prices,dates, period):
    alpha = 2 / (period + 1)
    ema = [(str(np.datetime_as_string(dates[0], unit='D')), prices[0])]
    for i in range(1, len(prices)):
        date = str(np.datetime_as_string(dates[i], unit='D'))
        ema_val = alpha * prices[i] + (1 - alpha) * ema[i-1][1]
        ema.append((date,ema_val))
    return ema
    
prices = df['close'].to_list()
dates = df.index.values

ema_lines = dict(alines= [calc_ema(prices,dates, 5), calc_ema(prices,dates, 10), calc_ema(prices,dates, 30) ], colors=['#ce17eb','#17a1eb', '#bdeb17'],linewidths=(0.5,0.5,0.5))
mpf.plot(df, type='candle', style='yahoo', volume=True, figratio=(8,4), alines = ema_lines)
```
